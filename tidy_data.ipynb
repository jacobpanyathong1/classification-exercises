```python
tidy_data.columns = ['name', 'Jan', 'Feb', 'Mar', 'April', 'May', 'Jun', 'Jul', 'Aug']
```

```python
tidy_data.head()
```

```python
tidy_data= tidy_data.melt(id_vars=['name'], var_name='month', value_name='day')
```

```python
tidy_data.replace({'P': 1, 'H': .5, 'T': .9, 'A': 0}, inplace=True)
```
