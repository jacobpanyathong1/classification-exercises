```python
def prep_telco(df):
    '''
     This function cleans the telco data to prepare for splitting
    '''
    df.drop(columns=['internet_service_type_id.1', 'payment_type_id.1'], inplace=True) #Drop duplicate columns
    encoded_vars = pd.get_dummies(df[['contract_type', 'internet_service_type', 'payment_type']])
    pd.set_option('display.max_columns', 500) # Display all columns
    df = pd.concat([telco_df, encoded_vars], axis=1)
    df['total_charges'] = (df.total_charges + '0').astype('float') # Change total charges to float
    df['total_charges'].dtype
    df['gender'] = df.gender.map({'Female': 1, 'Male': 0})
    df['partner'] = df.partner.map({'Yes': 1, 'No': 0})
    df['dependents'] = df.dependents.map({'Yes': 1, 'No': 0})
    df['phone_service'] = df.phone_service.map({'Yes': 1, 'No': 0})
    df['churn'] = df.churn.map({'Yes': 1, 'No': 0})
    return df
```

```python
train_iris_pw_versicolor = train_iris.loc[train_iris['species']== 'versicolor']['petal_width']
```

```python
train_iris_pw_setosa = train_iris.loc[train_iris['species']== 'setosa']['petal_width']
```

```python
stats.mannwhitneyu(train_iris_pw_virginica, train_iris_pw_versicolor) # Manwhy test for virginica and versicolor
```

```python
stats.mannwhitneyu(train_iris_pw_versicolor, train_iris_pw_setosa) # Manwhy test for versicolor and setosa
```

```python
stats.mannwhitneyu(x=train_iris_pw_virginica, y=train_iris_pw_setosa) # Manwhy test for virginica and setosa
```

**Takeaways**
$H_0$ : There is no siginifant difference between the species.
$H_a$ : There is no siginifant difference between the species.
With α = .05, our p-value is significantly below the α and with confidence we can reject the $H_0$ hypothesis that there is a significant difference between the species petal width.
